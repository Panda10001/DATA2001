```python
import geopandas as gpd
```

```python
countries_gdf = gpd.read_file("/Users/tiansuohaoer/Desktop/DATA/SA2_2021_AUST_SHP_GDA2020")
```

```python
countries_gdf
```

```python
countries_cleaned = countries_gdf[["SA2_CODE21","SA2_NAME21","AREASQKM21","geometry"]]
countries_cleaned
```

```python
countries_cleaned["wkt"] = countries_cleaned["geometry"].apply(lambda geom: geom.wkt if geom is not None else None)

countries_cleaned["geom_sql"] = countries_cleaned["wkt"].apply(
    lambda wkt: f"ST_GeomFromText('{wkt}', 4326)" if wkt is not None else None
)

countries_cleaned["geom_sql"]
```

```python

#  Connect to the Postgres database
conn = psycopg2.connect(
    dbname="postgres",
    user="postgres",
    password="114514",
    host="localhost",      # optional, default is localhost
    port="5432"            # optional, default is 5432
)
cur = conn.cursor()

# Loop through the GeoDataFrame and insert each row
for _, row in countries_cleaned.iterrows(): 
    sa2_code = row["SA2_CODE21"]
    name = row["SA2_NAME21"]
    geom_sql = row["geom_sql"]

    # Skip rows with missing geometry
    if geom_sql is None:
        print(f"Skipping row with missing geometry: {name}")
        continue

    # Safe SQL insert using parameter for name, but geom_sql is injected raw
    sql = f"""
    INSERT INTO spatial (sa2_code, name, geom)    
    VALUES (%s,%s, {geom_sql});                   
    """
    cur.execute(sql, (sa2_code , name))         

# Save changes and clean up
conn.commit()
cur.close()
conn.close()
```
